```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams["font.family"] = "sans-serif"
plt.rcParams["font.sans-serif"] = ["Microsoft YaHei"]
```

```python
tempfile = pd.read_csv("processed4basics.csv")
```

```python
tempfile
```

```python
tempfile.dropna(inplace=True)
tempfile
```

```python
tempfile.describe()
```

```python
tempfile.sort_values(by="粉丝数", ascending=False)
```

```python
tempfile["色温"] = tempfile["色温"][tempfile["色温"] > 0]
tempfile["色温"] = tempfile["色温"][tempfile["色温"] < 1e4]
```

```python
plt.figure(figsize=(8,8))
plt.scatter(tempfile["粉丝数"].to_numpy(), tempfile["色温"].to_numpy())
plt.xlabel("粉丝数")
plt.ylabel("色温")

plt.show()
plt.savefig('step2v1.png')
```

```python
sns.heatmap(tempfile.corr())
plt.savefig('step2v1_1.png')
```

```python
tempfile.corr()
```

```python
plt.figure(figsize=(10,8))
plt.xlim([0, 105000])
sns.regplot(x = "微博数",  
            y = "粉丝数",  
            data = tempfile,
            truncate=False)
pltXPosition = np.arange(0, 105000, 5000)
plt.xticks(pltXPosition, rotation=45)
#plt.xscale("log")
#plt.yscale("log")
plt.savefig('step2v1_2.png')
```
